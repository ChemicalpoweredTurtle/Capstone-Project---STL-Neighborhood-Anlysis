For the purposes of web scraping I am going to use the BeautifulSoup package. First we will import all libraries needed.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

Setup URL and download the web page

```python
url = 'https://www.bestplaces.net/find/zip.aspx?st=mo&msa=41180'
website_url = requests.get(url).text
```

Initialize and Print the BeautifulSoup Object. Looking at the webpage where I am pulling the data, there are a fair amount of issues. So I am going pull it in as text.

```python
soup = BeautifulSoup(website_url, 'html').text
soup
```

Now I am going to clean this text up. Should make everything a little easier.

```python
start_data = soup.find('All')
end_data = soup.find('\n \n\n\n')
soup_All=soup[start_data+8:end_data]
soup_All
```

Notice that all the zip codes start with a "6". That is going to make this very easy. I will just make a while loop to pull the data out of the string. 

```python
zipcode=[]
area=[]
while len(soup_All)>1:
    area_start=soup_All.find(' (')
    area_end=soup_All.find(')')
    area_name=soup_All[area_start+2:area_end]
    area.append(area_name)
    zipcode_start=soup_All.find('6')
    zipcode_end=soup_All.find(' ')
    zipcode.append(soup_All[zipcode_start:zipcode_end])
    soup_All=soup_All[area_end+1:len(soup_All)]
```

Now lets cast that into a dataframe!

```python
d= {'ZipCode': zipcode, 'AreaName': area}
df = pd.DataFrame(data=d)
df
```

```python
print('The final DataFrame shape is {}'.format(df.shape),'\n')
```

Now we will export this as a CSV file for later use!

```python
df.to_csv('STL.csv')
```
